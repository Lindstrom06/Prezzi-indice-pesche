# abbiamo inserito nan e fatto il grafico con alto, medio, basso

```python
def recode_cella(cella):
    if cella == ":":
        return np.nan
    else:
        return cella

indici_reali_pulito = indici_reali.applymap(lambda cella: recode_cella(cella))
```

```python
def recode_cella(cella):
    if cella > 110:
        return "Alto"
    elif cella <= 110 and cella > 90:
        return "Medio"
    elif cella <= 90:
        return "Basso"
    else:
        return np.nan
```

```python
recode_cella(np.nan)
```

```python
indici_reali_pulito = indici_reali_pulito[['2008', '2009', '2010', '2011', '2012', '2013', '2014','2015', '2016', '2017']].applymap(recode_cella)
```

```python
indici_reali_pulito["2008"].value_counts(dropna=False).plot(kind = "bar")
```
